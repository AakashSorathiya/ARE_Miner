```python
from transformers import pipeline
import torch
import pandas as pd
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
device
```

```python
review = 'Excellent running app with notifications and photos!'
```

```python
tfrex_model = 'quim-motger/t-frex-xlnet-base-cased'
classifier = pipeline("ner", model=tfrex_model, device=device)
```

```python
ner_results = classifier(review)
ner_results
```

```python
data = pd.read_csv('../datafiles/test_data_2.csv')
data.info()
```

```python
data2 = pd.read_csv('../datafiles/test_data.csv')
data2.info()
```

```python
features = []
for review in data2['Review'].to_list():
    entities = classifier(review)
    for entity in entities:
        features.append(entity['word'])
```
